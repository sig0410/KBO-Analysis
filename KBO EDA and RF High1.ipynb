### 자료들의 변동성을 보기 위한 절차 ###

```python
var='avg'#변동성이 큰 변수인 avg사용 
x=[]#그래프를 그리기 위해 x(1년전 avg),y(현재 avg)를 지정 
y=[]
reg1=reg.loc[reg['AB']>=100]
```

```python
reg1['batter_name'].unique()
#batter_name의 중복되지않은 이름 
```

```python
for name in reg1['batter_name'].unique(): 
    a=reg1.loc[reg1['batter_name']==name]
    k=[]
    for i in a['year'].unique():
        if (a['year']==i+1).sum()>0:#year에서 +1이 존재한다면
            k.append(i)
    for i in k:
        x.append(a.loc[a['year']==i,var].iloc[0])#1년전 avg
        y.append(a.loc[a['year']==i+1,var].iloc[0])#현재 avg  
#타석수가 100타석이상인 데이터를 reg1에 저장
#중복되지않는 타자들의 데이터수를 가지고 for문을 하여 a에 저장
#저장된 a에서 중복되지않은 year의 데이터길이를 이용하여 
#for를 사용해서 2016,2017,2018와 같은 연속적인 데이터들을 넣을수있다
#x=2016년데이터 y= 2017데이터 그 후 x=2017년 y=2018년 이렇게 되는 형태이다
```

### 코드가 길기때문에 함수화 ###

```python
def get_self_corr(var):
    x=[]
    y=[]
    reg1=reg.loc[reg['AB']>=100]

    for name in reg1['batter_name'].unique(): 
        a=reg1.loc[reg1['batter_name']==name]
        k=[]
        for i in a['year'].unique():
            if (a['year']==i+1).sum()>0:
                k.append(i)
        for i in k:
            x.append(a.loc[a['year']==i,var].iloc[0])#1년전 avg
            y.append(a.loc[a['year']==i+1,var].iloc[0])
    return(pd.Series(x).corr(pd.Series(y)))
#var의 상관계수를 리턴해준다 
```

### lag변수 ###

## Light GBM

트리형태의 발전형태 
트리에 비해 정교하게 에러값을 줄여줌
랜덤포레스트보다 성능이 좋음
결측값을 실수로 대체할 필요가 없음

```python
import lightgbm as lgb
param = {'num_leaves' : 100, 
         'min_data_in_leaf' : 30,
         'objective' : 'regression', # 모델의 목적 
         'max_depth' : 6,
         'learning_rate' : 0.02,
         'boosting' : 'gbdt',
         'feature_fraction' : 0.7522,
         'bagging_freq' : 1,
         'bagging_fraction' : 0.7083,
         'bagging_seed' : 11,
         'metric' : 'rmse', #평가지표 : rmse
         'random_state': 2019} 
# random_state : 모델이 만들어질때마다 모델의 정확도,예측값이 달라지는 경우가 많기때문에 고정해주는것

trn_data = lgb.Dataset(train[input_var], label=train['OPS'], weight=train['AB']) #가중치를 트레인의 타석으로 
#train[input_var] : x , label : y, weight : 가중치 
num_round = 10000
#800번을 업데이트하겠다
#light gbm은 트리를 여러개를 만드는데 트리를 하나씩 만들때 마다 예측치를 업데이트 
#그전의 예측치가 실제값과 얼마나 다른지를 비교하여 계속 업데이트하는 방식
#num_round가 많아지면 과대적합발생 일반성이 떨어짐  
clf = lgb.train(param, trn_data, num_round, valid_sets=[trn_data], verbose_eval=1000, early_stopping_rounds = 200)
#모델트레이닝한것을 clf에 넣은것
#verbose_eval : 100단위로 출력하겠다
#clf 트레이닝하는 과정 weight를 안해주면 그냥 rmse로 하지만 weight를 주면 clf에 적힌조건대로 수행
#valid_sets : 트레인데이터로 트레인할때마다 계속 예측치가 업데이트 되는데 
#그런 업데이트과정을 거칠때마다 새로운 예측값이 나오는데 그 예측값과 트레인의 OPS값을 비교해서 wrms을 뱉어주겠다 
#clf안에 lightgbm모델을 넣음
```

```python
predictions = clf.predict(test[input_var], num_iteration=clf.best_iteration)
#모델(2018년도까지로 만든 input_var)을 통해 예측을 하겠다
#테스트(2019년도)안의 input_var(lag_변수들과 total변수들)을 가지고 예측을 하겠다 
```
